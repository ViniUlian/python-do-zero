## Variáveis e tipos de dados simples

### Variáveis

Variáveis são rótulos que podem ser atribuidos para valores.

```python
mensagem = "Olá, Mundo!"
print(mensagem)
```

A variável pode ter o seu valor alterado e será mantido sempre o valor atual.

```python
mensagem = "Olá, Mundo!"
print(mensagem)
mensagem = "Tudo bem?"
print(mensagem)
```

Os nomes das variáveis precisam seguir algumas regras e boas práticas:
- Os nomes podem conter letras, números e underscores;
- Só podem começar com letras ou um underscore;
- Não pode haver espaço entre as palavras, para isso pode ser usado o underscore;
- Evitar usar nomes de funções Python e palavras reservadas nos nomes;
- Os nomes não precisam ser grandes, mas descritivos. Representar o que aquela variável significa. (Exemplo: usar nome ao invés de n);

Exemplos:

mensagem_1 (correto)

1_mensagem (errado)

nome_aluno (correto)

nome aluno (errado)

### Tipos de Dados

**String**

É uma sequência de caracteres. Tudo que estiver entre aspas é String.

```python
nome = "Sr Fulano"
print(nome)
```

É possível alterar as letras para maiúsculas e minúsculas utilizando métodos.

```python
nome = "jose silva"
print(nome.title()) # as letras iniciais de cada palavra ficam maiúsculas
```

```python
nome = "Jose Silva"
print(nome.lower()) # todas as letras ficam minúsculas
print(nome.upper()) # todas as letras ficam maiúsculas
```

É possível utilizar valores das variáveis dentro de Strings, com os f-strings, onde o Python formata (f) a String colocando os valores das variáveis.

```python
primeiro_nome = "jose"
segundo_nome = "silva"
print(f"{primeiro_nome} {segundo_nome}")
```

Para adicionar espaços em brancos nas strings podem ser utilizados os tabs (\t) ou as quebras de linhas (\n).

```python
print("Python")
print("\tPython")
```

```python
print("Linguagens:\nPython\nJava\nGo")
```

```python
print("Linguagens:\n\tPython\n\tJava\n\tGo")
```

Espaços em branco são considerados em Python. Então se o conteúdo for o mesmo, mas um tiver espaço, será considerado diferente: 'python' é diferente de 'python '.

rstrip(): retira o espaço em branco do lado direito da string. O resultado desse método é temporário, para que seja permanente é preciso associar o seu resultado a uma variável.

lstrip(): retira o espaço em branco do lado esquerdo da string. O resultado desse método é temporário, para que seja permanente é preciso associar o seu resultado a uma variável.

strip(): retira o espaço em branco de ambos os lados da string. O resultado desse método é temporário, para que seja permanente é preciso associar o seu resultado a uma variável.

```python
linguagem = " python "
print(f"|{linguagem}|")
print(f"|{linguagem.rstrip()}|")
print(f"|{linguagem.lstrip()}|")
print(f"|{linguagem.strip()}|")
```

Para retirar prefixos de uma string é preciso utilizar o método removeprefix('prefixo que quer retirar'). Esse método também não altera a string original.

```python
url = "https://meusite.com"
print(url.removeprefix('https://'))
```

**Números**

- Inteiros

```python
print(2 + 3)
print(3 - 2)
print(2 * 3)
print(3 / 2)
print(3 ** 2) # 3 ao quadrado
print(3 ** 3) # 3 ao cubo
print(2 + 3 * 4) # respeita a precedência das operações
```

- Floats

Números com um ponto decimal

```python
print(0.1 + 0.1)
print(0.2 + 0.2)
print(2 * 0.1)
```

O resultado de uma divisão será sempre um float, mesmo que seja a divisão de um inteiro por outro e que resulte em um inteiro

```python
8 / 4
```

Qualquer operação que tenha pelo menos um float, resultará em um float. Esse é o padrão do Python.

```python
print(1 + 2.0)
print(4 * 3.0)
print(5.0 ** 2)
```

É possível atribuir valores as varáveis de forma múltipla, ou seja, em uma única linha de código atribuir vários valores para várias variáveis. Para isso, a quantidade de variáveis tem que ser igual a dos valores

```python
x, y, z = 0, 0, 0
print(f"{x}, {y}, {z}")
```

```python
a, b, c = 4, 2.0, -1
print(f"{a}, {b}, {c}")
```

O Python não tem o tipo de dado constante integrado, mas para considerar uma variável constante, é preciso escrevê-la toda com letra maiúscula. Uma constante não deve ter seu valor alterado durante o programa.
